# MatplotLib or Pyplot
Its based on Matlab.
Each pyplot function makes some changes to figure like creates fig, creates plot area, decorates with lines and labels... 

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,1,6,3]);# y vals
plt.ylabel("My Plot"); 
plt.xlabel("Indices of y values"); #index of values in y
plt.show();
```

```python
plt.plot([1,2,3,4], [1,4,9,16]) #x and y axis
plt.ylabel("Squares"); 
plt.xlabel("Nums");
plt.grid();
plt.show();
```

```python
#'ro' is r->red o->circle or dot on Graph
plt.plot([1,2,3,4], [1,4,9,16],'ro') #x and y axis
plt.ylabel("Squares"); 
plt.xlabel("Nums");
plt.grid();
plt.show();
```

```python
import numpy as np
```

```python
#start end step size
t = np.arange(0.,5.,0.2)

#x-axis, y-axis, 'color Shape',label
#b-- ->blue dashed line
#rs ->red Squares
#g^ ->green triangles
#yo ->yellow circles
plt.plot(t,t**2,'b--',label='^2');
plt.plot(t,t**2.2,'rs',label='2.2');
plt.plot(t,t**2.5,'g^',label='2.5');
plt.plot(t,t**2.8,'yo',label='2.8');
plt.legend(); #to show symbol info
plt.grid();
plt.plot()
```

```python
x = [1,2,3,4];

y = [1,4,9,16];

colorShape = "b--";

plt.plot(x,y,colorShape,linewidth=5.0)
```

### To Plot Mutliple lines

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

#it returns 2 lines ->lines[0],lines[1]
lines = plt.plot(x1,y1,x2,y2);

#using setProperty method we can configure line
#its keyword args style of parameters
plt.setp(lines[0],color='b',linewidth=2.0);

#its matlab style of args
plt.setp(lines[1],'color','g',linewidth=5);

plt.grid();
```

# SubPlots /Multiple Plots/Graphs

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t);

# start ,end axis and step size
t1=np.arange(0.0,5.0,0.1);
t2=np.arange(0.0,5.0,0.3);

plt.figure(1);

# 211 indicates 2->2rows, 1->1col, 1->1st subplot
plt.subplot(211)
plt.grid();

#x x axis, y axis, colorShape
plt.plot(t1,f(t1), 'b--',label='t1');
plt.legend(); #line info at cornor

plt.subplot(212)
plt.grid();
plt.plot(t2,f(t2),'g^',label='t2');
plt.legend();

plt.show();
```

```python
plt.figure(1)

plt.subplot(211)
plt.plot([1,2,3])

plt.subplot(212)
plt.plot([4,5,6])



plt.figure(2)
plt.plot([4,5,6])



plt.figure(1) #the old figure 1 is used here
plt.subplot(211)
plt.title('Learning 1,2,3 in graphs')
plt.show()
```

### One figure 4 Subplots

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t);

t1=np.arange(0,1,0.2);
t2=np.arange(0,1,0.35);

plt.figure(1);

plt.subplot(221);
plt.plot(t1,f(t1),'ro',label='2rows 2cols 1stsubplot')
plt.legend();

plt.subplot(222);
plt.plot(t2,f(t2),'g^',label='2rows 2cols 2ndSubplot')
plt.legend();

plt.subplot(223);
plt.plot(t2,f(t2),'y--',label='2rows 2cols 3rdSubplot')
plt.legend();

plt.subplot(224);
plt.plot(t1,f(t1),'b+',label='2rows 2cols 4thSubplot');

plt.legend();
```
